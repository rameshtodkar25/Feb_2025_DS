```python
input='virat.kohli@rcb.com;rohit.sharma@mi.com;ms.dhoni@csk.com'
l1=input.split(';')
l1[0].index('.')
```

```python
input='virat.kohli@rcb.com;rohit.sharma@mi.com;ms.dhoni@csk.com'
l1=input.split(';')
fname,sname,cname=[],[],[]
d={}
for i in l1:
    first_dot=l1[0].index('.')
    second_dot=l1[0].index('.',first_dot+1)
    at_the_rate=l1[0].index('@')
    fname.append(i[:first_dot])
    sname.append(i[:first_dot+1:at_the_rate])
    cname.append(i[:at_the_rate+1:second_dot])

d['Fname']=fname
d['Sname']=sname
d['Cname']=cname
d
```
